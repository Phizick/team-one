```python
xls_path = "C:\work\ltc2022\data.xls"
dict_xls_exclude_sheets = set(["Паспорт"])
```

```python
f = pd.ExcelFile(xls_path)
sheet_names = set(f.sheet_names).difference(dict_xls_exclude_sheets)
sheets = pd.read_excel(
    io=f,
    skiprows=3,
    sheet_name=list(sheet_names),
    header=None,
    usecols=[0, 1, 2, 4, 5, 6, 7, 8, 9, 10],
    names=["unit_measure_class", "region_okato", "product_okpd2", "period",
           "2017", "2018", "2019", "2020", "2021", "2022"],
)
df = pd.concat(sheets.values())
df.reset_index(inplace=True)
```

```python
df = df[~df["period"].str.contains("январь-")]
```
